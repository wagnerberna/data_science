```python
import pandas
from matplotlib import pyplot, colors
# import matplotlib.pyplot as plot

# POPULAÇÃO:
data = pandas.read_csv("csv/04-classificados_tratado.csv", sep=",", encoding="iso-8859-1")
data.shape
```

```python
# AMOSTRAGEM: Cidade de Campinas
campinas = data.loc[data.NO_MUNICIPIO_RESIDENCIA == "Campinas"]
campinas.shape
```

```python
# HISTOGRAMA 
# bins número de colunas
campinas.IDADE.hist(bins=12)
pyplot.title("Distribuição de Idades - Enem Campinas 2019")
pyplot.show()
```

```python
campinas.RACA.hist(bins=5)
pyplot.title("Distribuição de Raças - Enem Campinas 2019")
pyplot.xlabel("Idade")
pyplot.ylabel("Frequência Absoluta")
pyplot.show()
```

```python
# rwidth (Espaçamento entre colunas)
# density ()
pyplot.hist(campinas.IDADE, 12, rwidth=1, color="red", density = True)
pyplot.title("Distribuição de Raças - Enem Campinas 2019")
pyplot.xlabel("Idade")
pyplot.ylabel("Frequência Relativa")
pyplot.show()
```

```python
pyplot.hist(campinas.IDADE, 12, rwidth=0.5)
pyplot.title("Distribuição de Raças - Enem Campinas 2019")
pyplot.xlabel("Idade")
pyplot.ylabel("Frequência Relativa")
pyplot.show()
```
