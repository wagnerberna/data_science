```python
import pandas
from plotly import express

# POPULAÇÃO:
data = pandas.read_csv("csv/04-classificados_tratado.csv", sep=",", encoding="iso-8859-1")
data.shape
```

```python
# AMOSTRAGEM: Cidade de Campinas
campinas = data.loc[data.NO_MUNICIPIO_RESIDENCIA == "Campinas"]
campinas.shape
```

```python
# Frequência Percentual Relativa
# (percent, probability, density)
grafico = express.histogram(campinas, x="RACA", nbins=5, histnorm="percent")
grafico.update_layout(width=500, height=500, title_text= "Distribuição de Raças Enem Campinas -Frequência Percentual Relativa")
grafico.show()
```

```python
# Frequência Relativa
grafico = express.histogram(campinas, x="RACA", nbins=5, histnorm="probability")
grafico.update_layout(width=500, height=500, title_text= "Distribuição de Raças Enem Campinas - Frequência Relativa")
grafico.show()
```

```python
# Frequência Absoluta
# Contagem
grafico = express.histogram(campinas, x="RACA", nbins=5, histnorm="density")
grafico.update_layout(width=500, height=500, title_text= "Distribuição de Raças Enem Campinas -Frequência Absoluta ", title_font_color = "Green")
grafico.show()
```

```python
# Frequência Absoluta
# DENSITY = Contagem
# COLOR coluna "SEXO" separa por SEXO Vermelho fem. e azul masc.
grafico = express.histogram(campinas, x="RACA", nbins=5, histnorm="density", color="SEXO")
grafico.update_layout(width=500, height=500, title_text= "Distribuição de Raças/Sexo Enem -Freq. Absoulta", title_font_color = "Green")
grafico.show()
```

```python
# Frequência Absoluta
# Contagem
# COLOR SEXO separa por SEXO Vermelho fem. e azul masc.
grafico = express.histogram(campinas, x="NOTA_MT", nbins=20, color="SEXO")
grafico.update_layout(width=500, height=500, title_text= "Distribuição por Sexo nota de Matemática-Frequência Absoluta", title_font_color = "red")
grafico.show()
```

```python
# Frequência Percentual Relativa
# Contagem
# COLOR SEXO separa por SEXO Vermelho fem. e azul masc.
grafico = express.histogram(campinas, x="NOTA_MT", nbins=20, histnorm="percent", color="SEXO")
grafico.update_layout(width=500, height=500, title_text= "Distribuição de sexo notas matemática -Frequência Percentual Relativa", title_font_color = "red")
grafico.show()
```

```python
# Frequência Absoluta
# Contagem
# COLOR SEXO separa por SEXO Vermelho fem. e azul masc.
grafico = express.histogram(campinas, x="NOTA_LC", nbins=20, histnorm="percent", color="SEXO")
grafico.update_layout(width=500, height=500, title_text= "Distribuição de Sexo Enem- Frequência Absoluta", title_font_color = "red")
grafico.show()
```
