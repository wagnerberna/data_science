```python
import pandas
import numpy
from plotly import express

# POPULAÇÃO:
data = pandas.read_csv("csv/04-classificados_tratado.csv", sep=",", encoding="iso-8859-1")
data.shape
```

```python
# MÉDIA geral idade
data["IDADE"].mean()
```

```python
data.IDADE.mean()
```

```python
round(data.IDADE.mean(),2)
```

```python
round(data.NOTA_REDACAO.mean())
```

```python
# MEDIANA IDADE GERAL (Valor do centro do array)
data.IDADE.median()
```

```python
data.NOTA_REDACAO.median()
```

```python
# Describe (Exibe o soma total, média, 50%->mediana, mínimo, máximo)
data.describe()
```

```python
round(data.describe(),1)
```

```python
# MODA (valor q mais se repete, com mais ocorrências)
data.IDADE.mode()
```

```python
round(data.NOTA_REDACAO.mean(), 1), data.NOTA_REDACAO.median(), data.NOTA_REDACAO.mode()
```

```python
grafico = express.histogram(data, x="NOTA_REDACAO", nbins=15)
grafico.update_layout(width=400, height=400, title_text= "Nota Redação - ", title_font_color = "grey")
grafico.show()
```

```python
grafico = express.histogram(data, x="NOTA_REDACAO", nbins=15, histnorm="percent", color="SEXO")
grafico.update_layout(width=400, height=400, title_text= "Distribuição de Raças Enem -Frequência Percentual Relativa", title_font_color = "grey")
grafico.show()
```
