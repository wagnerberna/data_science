```python
import pandas
from seaborn import displot, histplot

# POPULAÇÃO:
data = pandas.read_csv("csv/04-classificados_tratado.csv", sep=",", encoding="iso-8859-1")
data.shape

```

```python
# AMOSTRAGEM: Cidade de Campinas
campinas = data.loc[data.NO_MUNICIPIO_RESIDENCIA == "Campinas"]
campinas.shape
```

```python
displot(campinas["RACA"], bins=5, color="brown", kde=False)
```

```python
# ";" no final para não aparecer mensagem
histplot(campinas["RACA"], bins=5, color="brown", kde=False);
```

```python
# kde=True Linha de densidade como é distribuído o histograma
histplot(campinas["RACA"], bins=5, color="brown", kde=True);
```

```python
# DEFINE A COLUNA PELO x
# probability soma igual 1
# (probability, frequency, density, count)
histplot(campinas, x="RACA", bins=5, color="orange", kde=False, stat="probability");
```

```python
# frequency absoluta
# (probability, frequency, density, count)
histplot(campinas, x="RACA", bins=5, color="orange", kde=False, stat="frequency");
```

```python
# frequency absoluta
# (probability, frequency, density, count)
histplot(campinas, x="RACA", bins=5, color="orange", kde=False, stat="count");
```
