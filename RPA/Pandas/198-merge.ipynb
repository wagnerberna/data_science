```python
import pandas as pd
```

```python
df_notas1 = pd.DataFrame({
    "Nome": ["Diana", "Clark", "Bruce", "Arthur"],
    "Port":[9, 7, 8, 5],
    "Mat":[7, 5, 9, 4],
})
df_notas1
```

```python
df_notas2 = pd.DataFrame({
    "Bio":[0, 9, 8, 7],
    "Hist":[10, 2, 9, 5],
    "Nome": ["Diana", "Clark", "Bruce", "Barry"],
})
df_notas2
```

#### Merge - Inner
-on: nome da coluna igual em ambas

-how: como fazer o merge

-inner: junta SOMENTE as chaves q existirem EM AMBAS tabelas

```python
df_merge_notas1 = df_notas1.merge(df_notas2)
df_merge_notas1
```

```python
# Inner
df_merge_notas2 = pd.merge(df_notas1, df_notas2, on=["Nome"], how="inner")
df_merge_notas2
```

```python
# definindo colunas de cada coluna que serão incluídas
# suffixes ???
df_merge_notas_resumo1 = pd.merge(df_notas1[["Nome", "Port"]], df_notas2[["Nome", "Hist"]], on=["Nome"], how="inner", suffixes=("Notas 1", "Notas 2"))
df_merge_notas_resumo1
```

#### Merge - Left

```python
# left
df_merge_notas3 = pd.merge(df_notas1, df_notas2, on=["Nome"], how="left")
df_merge_notas3
```

#### Merge - Outer
-Junta o conteúdo de TODAS tabelas

```python
# outer
df_merge_notas4 = pd.merge(df_notas1, df_notas2, on=["Nome"], how="outer")
df_merge_notas4
```
