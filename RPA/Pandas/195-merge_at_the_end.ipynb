```python
import pandas as pd
```

```python
df_notas1 = pd.DataFrame({
    "Nome": ["Diana", "Clark", "Bruce"],
    "Port":[9, 7, 8],
    "Mat":[7, 7, 9],
    "Bio":[7, 7, 7]
})
df_notas1
```

```python
df_notas2 = pd.DataFrame({
    "Nome": ["Barry", "Arthur"],
    "Port":[9, 7],
    "Mat":[7, 7],
    "Bio":[7, 7]
})
df_notas2
```

#### Merge planilhas com colunas iguais add no final
-Concat

```python
df_notas_geral = pd.concat([df_notas1, df_notas2])
df_notas_geral
```

```python
df_notas_geral_grupo = pd.concat([df_notas1, df_notas2], keys=["Notas1", "Notas2"])
df_notas_geral_grupo
```

```python
df_resume = df_notas_geral[["Nome", "Port", "Mat"]]
df_resume
```
