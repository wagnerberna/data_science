```python
import pandas as pd
import numpy as np
```

```python
df_notas = pd.DataFrame({
    "Nome": ["Diana", "Clark", "Bruce", "Barry", np.nan, np.nan],
    "Port":[7, 5, 8, 9, np.nan, np.nan],
    "Mat":[5, np.nan, 9, 9, np.nan, np.nan],
    "Bio":[8, np.nan, 8, 5, np.nan, np.nan]
})
df_notas
```

```python
# Remover linhas em branco pela coluna  "subset"
df_notas2 = df_notas.dropna(subset=["Nome"]).reset_index(drop=True)
df_notas2
```

```python
# Deleta todas as linhas q tem ao menos 1 valor vazio
df_notas3 = df_notas.dropna()
df_notas3
```

```python
df_notas4 = pd.DataFrame({
    "Nome": ["Diana", "Clark", "Bruce", "Barry", np.nan, np.nan],
    "Port":[7, 5, 8, 9, np.nan, np.nan],
    "Mat":[5, np.nan, 9, 9, np.nan, np.nan],
    "Bio":[8, np.nan, 8, 5, np.nan, np.nan]
})

df_notas5 = pd.DataFrame({
    "Nome": ["Diana", "Clark", "Bruce", "Barry", np.nan, np.nan],
    "Port":[7, 5, 8, 9, np.nan, np.nan],
    "Mat":[5, np.nan, 9, 9, np.nan, np.nan],
    "Bio":[8, np.nan, 8, 5, np.nan, np.nan]
})
df_notas4
```

```python
# Preencher valores faltantes da coluna com a m√©dia
df_notas4["Bio"] = df_notas4["Bio"].fillna(df_notas4["Bio"].mean())
df_notas4["Mat"] = df_notas4["Mat"].fillna(0)
df_notas4["Port"] = df_notas4["Mat"].fillna(0)

df_notas4
```

```python
# ffill preenche com o valor 
df_notas5["Bio"] = df_notas5["Bio"].ffill()
df_notas5["Port"] = df_notas5["Mat"].ffill()
df_notas5["Mat"] = df_notas5["Mat"].ffill()

df_notas5
```
