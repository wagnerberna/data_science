```python
import pandas as pd
```

```python
df_notas1 = pd.DataFrame({
    "Nome": ["Diana", "Clark", "Bruce", "Arthur", "Barry"],
    "Apelido": ["Mulher Maravilha", "Superman", "Batman", "Aquaman", "Flash"],
    "Poder": ["Força", "Força", "Estratégia", "Força", "Velocidade"],
    "Nível Força":[8, 10, 6, 8, 6],
    "Nível Velocidade":[8, 9, 5, 8, 10],
})
df_notas1
```

#### Pivot

```python
df_pivot = df_notas1.pivot_table(index="Poder", columns="Nome", values="Nível Força")
df_pivot
```

```python
df_pivot = df_notas1.pivot_table(index="Poder", columns="Nome", values=["Nível Força", "Nível Velocidade"])
df_pivot
```

```python
df_pivot = df_notas1.pivot_table(index="Poder", columns=["Nome", "Apelido"], values=["Nível Força", "Nível Velocidade"].fill)
df_pivot
```
