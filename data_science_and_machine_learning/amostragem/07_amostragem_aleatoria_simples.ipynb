```python
# Amostra Aleatória Simples
# random_state: fixa o mesmo resultado baseado no Seed passado.
# Seed: número semente geradora aleatória
df_simple_random_sample = dataset.sample(n=100, random_state=1)
df_simple_random_sample.shape
```

```python
# Criar fn de coleta aleatória
def simple_random_sample(dataset, sample_quantity):
    return dataset.sample(n=sample_quantity, random_state=1)
```

```python
df_simple_random_sample = simple_random_sample(dataset, 100)
df_simple_random_sample.shape
```
