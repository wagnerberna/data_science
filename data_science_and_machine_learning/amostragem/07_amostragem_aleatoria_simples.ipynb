```python
# Criar fn de coleta aleat√≥ria
def simple_random_sample(dataset, sample_quantity):
    return dataset.sample(n=sample_quantity, random_state=1)
```

```python
df_simple_random_sample = simple_random_sample(dataset, 100)
df_simple_random_sample.shape
```
