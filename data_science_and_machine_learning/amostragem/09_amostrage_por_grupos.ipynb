```python
import pandas as pd
import random
import numpy as np

file_census = "../data_base/census.csv"

dataset = pd.read_csv(file_census)

# Indica o n√∫mero de Linhas e colunas
dataset.shape
```

```python
# Amostragem por Grupos
populacao = len(dataset)
numero_de_grupos = 10
quantidade_por_grupo = populacao / numero_de_grupos
quantidade_por_grupo
```

```python
# for "_": n√£o cria vari√°vel para controlar o for
# iterrows: vai percorrer cada um dos registros da base de dados
grupos = []
id_grupo = 0
contagem = 0
for _ in dataset.iterrows():
    grupos.append(id_grupo)
    contagem += 1
    if contagem > 3256:
        contagem = 0
        id_grupo += 1

print(grupos)
```

```python
def sample_group(dataset, group_quantity):
    population_quantity = len(dataset)
    interval = round(population_quantity / group_quantity)
    # random.seed(1)
    group_list = []
    id_group = 0
    count = 0
    
    for el in dataset.iterrows():
        group_list.append(id_group)
        count += 1
        if count > interval:
            count = 0
            id_group += 1
    
    dataset["grupo"] = group_list
    selected_group_code = random.randint(0, group_quantity)
    selected_group = dataset[dataset["grupo"] == selected_group_code]
    return selected_group      
```

```python
df_sample_group = sample_group(dataset, 100)
df_sample_group.shape, df_sample_group["grupo"].value_counts()
```
