```python
import pandas as pd


df = pd.read_excel('ibge.xlsx', 'Tab 1.1.38')

colNm = [
         'Regiao',
          'MedioTotal',
          'MedioUtilizaram',
          'MedioNaoUtilizaram',
          'MedianoTotal',
          'MedianoUtilizaram',          
          'MedianoNaoUtilizaram',
          ]

df.columns = colNm

df.head()
```

```python
df = df.drop([0, 1 ,2 ,3 , 46, 47])

df = df.set_index(df['Regiao'])

df = df.drop(columns=['Regiao'])

df.head()
```

```python
df = df.drop(['Brasil'])

df = df.sort_values('MedioTotal')

df.head()
```

```python
df.plot(rot=90)
```

```python
# Converte todos dados para numérico
df = df.apply(pd.to_numeric, errors='coerce')

# faz a correlação dos dados
matCorr = df.corr()

# pd.figure.clear()
df.boxplot(rot=90)
```
