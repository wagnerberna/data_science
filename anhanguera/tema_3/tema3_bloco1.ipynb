```python
import numpy as np
import random
from matplotlib import pyplot as plt
import pandas

x = range(1, 50)
y = [random.random()  * 100 for i in x]

print(x, y)
```

```python
plt.scatter(x, y)
```

```python
plt.bar(x, y)
```

```python
plt.plot(x, y, color = "red", linestyle="dashed")
plt.grid(True)
plt.legend(["Dados Aleatórios"])
plt.title("Gráfico Teste")
plt.xlabel("Eixo X")
plt.ylabel("Eixo Y")
```
